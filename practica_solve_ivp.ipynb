```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la EDO: y' = f(t, y)

def f(t, y):
    return -y
```

```python
# Definir intervalos de integración

t0, tf = 0.0, 0.5 # Intervalo de integración 
y0 = [1.0] # condición inical (lista o array)

t_eval = np.linspace(t0, tf, 20)
```

```python
# Resolver el PVI con solve_ipv (RK45 por defecto = Runge- Kutta)

sol = solve_ivp(
    fun=f,          # función del sistema
    t_span=(t0,tf), # intervalo de tiempo.
    y0=y0,          # condición inicial.
    method='RK45',  # usa Runge–Kutta 4(5) con paso adaptativo (RKF45)
    t_eval=t_eval   # pide la solución exactamente en esos 200 puntos.
)
```

```python
# Extraer la solución

t = sol.t  # tiempos
y_num = sol.y[0]   # solución numperica
```

```python
# Solución exacta para comparar

y_exact = np.exp(-t)
```

```python
# Graficar con Matplotlib

fig = plt.figure(figsize=(6,4), dpi=150)
ax = fig.add_subplot(111)

ax.plot(t, y_exact, 'k-', label='Exacta $e^{-t}$' )
ax.plot(t, y_num, 'ro', label='Numérica RK45', markersize=6)

ax.set_xlabel('$t$', fontsize=14)
ax.set_ylabel('y(t)', fontsize=14)
ax.legend(loc='best', fontsize=10)
ax.tick_params(axis='both', which='major', labelsize=10)
plt.show()
```
