# Linear Algebra

## Eigenvalue problems

Eigenvalue problems are fundamental concepts in linear algebra and have applications in various fields such as physics, engineering, and computer science. An eigenvalue problem involves finding the eigenvalues and associated eigenvectors of a square matrix.

An eigenvalue problem involves the following equation:


\begin{equation}\label{eq:eigenval_prob}
\mathbf{A x}=\lambda \mathbf{x}
\end{equation}

where:

<ul style="text-align:justify">
    <li>
        $\textbf{A}$:  $n \times n$ known square matrix
    </li>
    <li>
        $\lambda$ : unknown scalar
    </li>
    <li>
        $x$: unknown vector
    </li>
</ul>



In the previous problem:

<ul>
    <li>  All numerical values of $\lambda$ that satisfy (\ref{eq:eigenval_prob}) are named <b>eigenvalues</b></li>
    <li>  The corresponding vectors $x\neq 0$ for each $\lambda$ are named <b>eigenvectors</b></li>
</ul>

<p style="font-style:italic">Notice that this problem involves finding <b>eigenvectors</b> $\textbf{x}$, such that when multiplying $\textbf{A}$, they have the same effect than multiplying $\textbf{A}$ by the scalar $\lambda$.</p>

### Finding Eigenvalues and Eigenvectors

To find the eigenvalues and eigenvectors of a matrix A, we can rearrange (\ref{eq:eigenval_prob}) to obtain (\ref{eq:eigenval_prob_char}):

\begin{equation}\label{eq:eigenval_prob_char}
(\mathbf{A}-\lambda \mathbf{I}) \mathbf{x}=\mathbf{0}  
\end{equation}

which is the same as:

$$
\left[\begin{array}{cccc}
a_{11}-\lambda & a_{12} & \cdots & a_{1 n} \\
a_{21} & a_{22-\lambda} & \cdots & a_{2 n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{n 1} & a_{n 2} & \cdots & a_{n n-\lambda}
\end{array}\right]\left[\begin{array}{c}
x_{1} \\
x_{2} \\
\vdots \\
x_{n}
\end{array}\right]=\left[\begin{array}{c}
0 \\
0 \\
\vdots \\
0
\end{array}\right]
$$

and is a <b>system of linear equations</b>, particularly an homogeneous one.

An homogeneous system of linear equations:

\begin{equation}
\mathbf{A x}=\mathbf{0}
\end{equation}

will have an unique solution (trivial)

$$\textbf{x} = \left[\begin{array}{c}
0 \\
0 \\
\vdots \\
0
\end{array}\right]$$

if:

$$
\operatorname{det} \mathbf{A} \neq 0
$$

Instead, it will have non-trivial solutions <b>if and only if</b>:

$$
\operatorname{det} \mathbf{A}=0
$$

<b> <p style="color:red"> Accordingly, if we want to find non-trivial solutions to the eigenvalue problem, we must ensure that $\operatorname{det} \mathbf{A}=0$</p></b> Therefore, our eigenvalue problem (\ref{eq:eigenval_prob}) will have a solution if:

\begin{equation}\label{eq:eigenval_prob_char2}
\operatorname{det}(\mathbf{A}-\lambda \mathbf{I})=0
\end{equation}

<ul>
    <li>
        equation (\ref{eq:eigenval_prob_char2}) is known as the <b>characteristic equation</b>.
    </li>
    <li>
        $(\mathbf{A}-\lambda \mathbf{I})$ is known as the characteristic matrix
    </li>
    <li>
        When expanding the characteristic determinant, the obtained polynomial is known as the characteristic determinant.
    </li>
</ul>

Finally, in order to compute the eigenvalues and eigenvectors of $\textbf{A}$, we have to solve the characteristic equation (\ref{eq:eigenval_prob_char2}).

### Computing eigenvalues and eigenvectors using scipy (through an example)

<a> https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.eig.html </a>

let the matrix $\mathbf{A}$ be:

$$
\mathbf{A}=\left[\begin{array}{ccc}
13 & 5 & 2 \\
2 & 7 & -8 \\
5 & 4 & 7
\end{array}\right]
$$

we will compute its eigenvalues and eigenvectors using the the <b>eig</b> function of the <b>scipy.linalg</b> module

```python
import numpy as np
from scipy.linalg import eig

# the following modules are used to print a string in latex equation format
from IPython.display import Math, display
def p_print(text):
    display(Math(text))
```

```python
A = np.array([[13,5,2],[2,7,-8],[5,4,7]])
A
```

```python
eig_val, eig_vect = eig(A)
```

Notice that:

<ul>
    <li>
        the <b>eig</b> function returns an array with two elements: eigenvalues and eigenvectors. Here, we inmedatly splited these results into two arrays for convenience.
    </li>
    <li>
        Eigenvalues and eigenvectors are returned as complex values (which is a common occurence!)
    </li>
    <li>
        Eigenvectors are returned as columns of the array containing them
    </li>
</ul>       

```python
for i,v in enumerate(eig_val):
    p_print(fr'\lambda_{i+1}={v}')
```

```python
for i,v in enumerate(eig_vect[1,:]):
    p_print(fr'e_{i+1} = {list(eig_vect[:,i])}')
```
