### Using f-strings

f-strings provide a convenient and readable way to format strings by embedding expressions directly within them. This makes it easy to create dynamic strings that include variables, calculations, and other Python expressions.

To create an f-string, prepend the string with an 'f' or 'F' and enclose expressions within curly braces {}.

```python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

You can use any valid Python expression inside the curly braces. This allows you to include calculations, functions, and more.

```python
length = 5
width = 3
area = length * width
print(f"The area of the rectangle is {area} square units.")
```

You can use f-strings to control the formatting of numbers, including specifying decimal places and alignment.

```python
pi = 3.14159265
print(f"The value of pi is approximately {pi:.2f}")
print(f"The value of pi is approximately {pi:20.5e}")
```

## Control Flow

Control flow structures allow you to control the execution of your code based on conditions. This is essential for creating more dynamic and interactive programs.

### Conditional Statements

Conditional statements are used to execute different blocks of code based on whether a condition is true or false.

- The <b>if</b> statement checks a condition and executes the code inside the block if the condition is true.
- The <b>elif</b> (short for "else if") statement is used for additional conditions.
- The <b>else</b> statement is executed if none of the preceding conditions are true.

#### Example

```python
age = 18
if age < 18:
    print("You are underage.")
elif age == 18:
    print("You just turned 18!")
else:
    print("You are an adult.")
```

#### Conditional Statements using Booleans

When using booleans prefer the following syntax:

```python
test_var = True

if test_var:
    print('its True')
```

Instead of:

```python
if test_var == True:
    print('its True')
```

Notice that in Python, non-zero floats are considered 'True' booleans, whereas 0 is considered 'False'.

```python
if not 0:
    print('0 is False')
```

```python
x=-1
if x:
    print('x is not False')
```

### Loops

Loops allow you to repeat a block of code multiple times.

- The <b>for</b> loop iterates over a sequence (e.g., a list, range) and performs actions for each element.
- The <b>while</b> loop repeats a block of code as long as a specified condition is true.
Example (for loop):

```python
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print("I love", fruit)
```

```python
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```

## Functions

Functions are blocks of code that can be called and executed multiple times with different inputs. They allow you to organize your code into reusable units and make it more modular.

### Defining Functions:

To define a function, use the def keyword followed by the function name, parameter list, and a colon :. The indented block following the colon is the function body.

```python
def greet(name):
    print(f"Hello, {name}!")
```

### Function Parameters:

Functions can take parameters, which are inputs provided when the function is called. Parameters are specified in the function's definition.

```python
def add_numbers(a, b):
    return a + b
```

### Return Statements:

Functions can use the return statement to send a result back to the caller. This allows you to use the result of a function in other parts of your code.

```python
def square(x):
    return x ** 2
```

### Function Call:

To use a function, call it by its name and provide the required arguments. The function's code block will be executed.

```python
greet("Alice")
sum_result = add_numbers(5, 3)
squared_value = square(4)
```

### Default Parameters:

You can provide default values for parameters in case they are not explicitly provided when the function is called.

```python
def power(base, exponent=2):
    return base ** exponent

power(2)
```

## File Handling

File handling is essential for reading and writing data to files on your computer. This is useful for various tasks, such as reading input data, writing output, and persisting data between program runs.

### Opening Files

To open a file, you use the built-in open() function. You provide the file's name and specify whether you want to read, write, or append to the file.

```python
# writing
file = open("data.txt", "w")
file.write("Hello, world!")
file.close()
```

```python
# reading
file = open("data.txt", "r")
content = file.read()
file.close()
```

### Context Managers (with Statement)

Using context managers (the with statement) is a safer way to work with files. It ensures that the file is properly closed, even if an exception is raised.

```python
with open("data.txt", "r") as file:
    content = file.read()
```

### Reading Lines

You can read a file line by line using a loop.

```python
with open("data.txt", "r") as file:
    for line in file:
        print(line.strip())  # Strip removes leading/trailing whitespace
```

## Challenge: Counting letters

Extracted from <a>https://projecteuler.net/problem=17</a>

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are $3+3+5+4+4=19$ letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

<b>NOTE</b>: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
# Challenge: Counting letters
# Solution as a beginner student

# First, let's create dictionaries for the basic number words

# Numbers from 1 to 19
ones = {
    1: "one", 2: "two", 3: "three", 4: "four", 5: "five",
    6: "six", 7: "seven", 8: "eight", 9: "nine", 10: "ten",
    11: "eleven", 12: "twelve", 13: "thirteen", 14: "fourteen",
    15: "fifteen", 16: "sixteen", 17: "seventeen", 18: "eighteen", 19: "nineteen"
}

# Tens (20, 30, 40, etc.)
tens = {
    20: "twenty", 30: "thirty", 40: "forty", 50: "fifty",
    60: "sixty", 70: "seventy", 80: "eighty", 90: "ninety"
}

# Function to convert a number to words
def number_to_words(n):
    # Special case for 1000
    if n == 1000:
        return "one thousand"
    
    # Numbers 1-19
    if n in ones:
        return ones[n]
    
    # Numbers 20-99
    if n < 100:
        ten_digit = (n // 10) * 10  # Get the tens (20, 30, 40...)
        one_digit = n % 10  # Get the ones digit
        if one_digit == 0:
            return tens[ten_digit]
        else:
            return tens[ten_digit] + "-" + ones[one_digit]
    
    # Numbers 100-999
    if n < 1000:
        hundred_digit = n // 100  # Get the hundreds digit
        remainder = n % 100  # Get the remainder
        
        if remainder == 0:
            return ones[hundred_digit] + " hundred"
        else:
            # British usage: use "and" between hundred and remainder
            return ones[hundred_digit] + " hundred and " + number_to_words(remainder)

# Function to count letters (no spaces or hyphens)
def count_letters(word):
    count = 0
    for char in word:
        # Only count letters, not spaces or hyphens
        if char != " " and char != "-":
            count = count + 1
    return count

# Now let's count all letters from 1 to 1000
total_letters = 0

for number in range(1, 1001):
    # Convert number to words
    word = number_to_words(number)
    # Count the letters
    letter_count = count_letters(word)
    # Add to total
    total_letters = total_letters + letter_count

print(f"Total letters from 1 to 1000: {total_letters}")

# --- Check examples: change these numbers to test ---
examples_to_check = [1, 2, 315, 4, 5, 115, 342]

for n in examples_to_check:
    w = number_to_words(n)
    c = count_letters(w)
    print(f"{n} = {w} -> {c} letters")
```
