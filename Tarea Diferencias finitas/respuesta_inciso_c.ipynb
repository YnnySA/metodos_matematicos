##### Esquemas de diferencias finitas con paso variable

En la conferencia se define el paso de integración como $\Delta x_n = x_{n+1} - x_n$. Cuando $\Delta x_n$ **no** es constante, las fórmulas estándar dejan de ser aplicables sin modificación. Partiendo de la **expansión de Taylor** alrededor de $x_n$ con pasos a izquierda y derecha distintos:

$$h_1 = x_n - x_{n-1}, \qquad h_2 = x_{n+1} - x_n,$$

se obtienen los siguientes esquemas adaptados:

- **Diferencia progresiva** ($\mathcal{O}(h_2)$):
$$f'_n \approx \frac{f_{n+1} - f_n}{x_{n+1} - x_n} = \frac{f_{n+1} - f_n}{h_2}$$

- **Diferencia regresiva** ($\mathcal{O}(h_1)$):
$$f'_n \approx \frac{f_n - f_{n-1}}{x_n - x_{n-1}} = \frac{f_n - f_{n-1}}{h_1}$$

- **Diferencia central de 3 puntos no uniformes** ($\mathcal{O}(h_1\, h_2)$):

Derivando el polinomio interpolante de grado 2 que pasa por $(x_{n-1}, f_{n-1})$, $(x_n, f_n)$, $(x_{n+1}, f_{n+1})$:

$$\boxed{\,f'_n \approx \frac{-h_2}{h_1\,(h_1+h_2)}\,f_{n-1} + \frac{h_2 - h_1}{h_1\,h_2}\,f_n + \frac{h_1}{h_2\,(h_1+h_2)}\,f_{n+1}\,}$$

Notar que cuando $h_1 = h_2 = h$ esta expresión se reduce a la diferencia central clásica $\frac{f_{n+1}-f_{n-1}}{2h}$.


```python
def dif_progresiva_nu(x, y):
    """Diferencia progresiva para puntos no uniformes (x, y)"""
    df = np.zeros_like(y)
    # Progresiva en todos excepto el último punto
    df[:-1] = (y[1:] - y[:-1]) / (x[1:] - x[:-1])
    # En el último punto: regresiva
    df[-1]  = (y[-1] - y[-2]) / (x[-1] - x[-2])
    return df


def dif_regresiva_nu(x, y):
    """Diferencia regresiva para puntos no uniformes (x, y)"""
    df = np.zeros_like(y)
    # Regresiva en todos excepto el primer punto
    df[1:] = (y[1:] - y[:-1]) / (x[1:] - x[:-1])
    # En el primer punto: progresiva
    df[0]  = (y[1]  - y[0])  / (x[1]  - x[0])
    return df


def dif_centrada_nu(x, y):
    """Diferencia central de 3 puntos para malla no uniforme.

    Obtenida derivando el polinomio interpolante de Lagrange de grado 2.
    Coincide con la diferencia central clásica si h_1 = h_2.
    """
    df = np.zeros_like(y)
    # Pasos a izquierda (h1) y derecha (h2) en los puntos interiores
    h1 = x[1:-1] - x[:-2]      # x_n - x_{n-1}
    h2 = x[2:]   - x[1:-1]     # x_{n+1} - x_n

    # Coeficientes del esquema centrado no uniforme
    c_m1 =  -h2          / (h1 * (h1 + h2))   # coef. de f_{n-1}
    c_0  =  (h2 - h1)    / (h1 * h2)          # coef. de f_n
    c_p1 =   h1          / (h2 * (h1 + h2))   # coef. de f_{n+1}

    df[1:-1] = c_m1 * y[:-2] + c_0 * y[1:-1] + c_p1 * y[2:]

    # En los extremos usamos progresiva / regresiva
    df[0]  = (y[1]  - y[0])  / (x[1]  - x[0])
    df[-1] = (y[-1] - y[-2]) / (x[-1] - x[-2])
    return df
```

##### Evaluación de la primera derivada

Aplicamos los tres esquemas a la nube de puntos no homogénea y graficamos los resultados de forma conjunta.


```python
# Cálculo de la primera derivada con los tres esquemas no uniformes
df_prog = dif_progresiva_nu(x, y)
df_regr = dif_regresiva_nu(x, y)
df_cent = dif_centrada_nu(x, y)

# Graficación comparativa
fig, ax = plt.subplots(figsize=(7, 4), dpi=150)

ax.plot(x, df_prog, 'C1-',  linewidth=0.7, label='Progresiva  $\\mathcal{O}(h_2)$')
ax.plot(x, df_prog, 'C1.', markersize=3)

ax.plot(x, df_regr, 'C2-',  linewidth=0.7, label='Regresiva  $\\mathcal{O}(h_1)$')
ax.plot(x, df_regr, 'C2.', markersize=3)

ax.plot(x, df_cent, 'C3-',  linewidth=0.9, label='Centrada no uniforme  $\\mathcal{O}(h_1 h_2)$')
ax.plot(x, df_cent, 'C3.', markersize=3)

ax.set_xlabel('x', fontsize=11)
ax.set_ylabel("$dy/dx$", fontsize=11)
ax.set_title("Primera derivada por diferencias finitas (espaciado no homogéneo)", fontsize=11)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8, loc='best')
plt.tight_layout()
plt.show()
```

##### Comparación por separado

Se muestran los tres esquemas en una grilla $1 \times 3$, tomando la **diferencia central no uniforme** como referencia visual ya que su orden de error es superior al de las diferencias progresiva y regresiva.


```python
# Comparación contra la centrada no uniforme (referencia)
fig, axes = plt.subplots(1, 3, figsize=(11, 3.5), dpi=150, sharey=True)

esquemas = [
    ('Progresiva  $\\mathcal{O}(h_2)$',                 df_prog, 'C1'),
    ('Regresiva  $\\mathcal{O}(h_1)$',                  df_regr, 'C2'),
    ('Centrada no uniforme  $\\mathcal{O}(h_1 h_2)$',   df_cent, 'C3'),
]

for ax, (nombre, df, color) in zip(axes, esquemas):
    ax.plot(x, df_cent, 'k-', linewidth=0.8, alpha=0.4, label='Ref. centrada n.u.')
    ax.plot(x, df, f'{color}-', linewidth=0.7, label=nombre)
    ax.plot(x, df, f'{color}.', markersize=3)
    ax.set_title(nombre, fontsize=10)
    ax.set_xlabel('x', fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=7, loc='best')

axes[0].set_ylabel('$dy/dx$', fontsize=10)
plt.tight_layout()
plt.show()
```

##### Comparación de errores entre esquemas

Como no se conoce la función analítica $f(x)$, se utiliza la **diferencia centrada no uniforme** como referencia (su error de truncamiento es $\mathcal{O}(h_1 h_2)$, mientras que las diferencias progresiva/regresiva son $\mathcal{O}(h)$). Se calcula entonces:

$$E_i = \left| f'_{\text{esquema}}(x_i) - f'_{\text{central n.u.}}(x_i) \right|$$

y se reportan el **error máximo** y el **error promedio** sobre los puntos interiores ($i = 1, \ldots, N-2$), donde los tres esquemas usan sus fórmulas nominales.


```python
# Errores respecto a la centrada no uniforme, sobre los puntos interiores
mask = np.zeros_like(x, dtype=bool)
mask[1:-1] = True

ref = df_cent
errores = {
    'Progresiva': df_prog,
    'Regresiva':  df_regr,
}

print(f"{'Esquema':<15}{'Error máx.':>16}{'Error prom.':>16}")
print('-' * 47)
for nombre, df in errores.items():
    err = np.abs(df[mask] - ref[mask])
    print(f"{nombre:<15}{err.max():>16.5e}{err.mean():>16.5e}")

# Información adicional sobre el rango de pasos
p_print(fr'\Delta x_{{\min}} = {np.diff(x).min():.5f}, \quad \Delta x_{{\max}} = {np.diff(x).max():.5f}')
```

##### Análisis de los resultados

A partir de los gráficos y de la tabla de errores se observan los siguientes comportamientos:

1. **Necesidad de adaptar los esquemas.** A diferencia del inciso b), donde $h$ era constante, en este caso $\Delta x_i$ varía entre $\Delta x_{\min}$ y $\Delta x_{\max}$. Aplicar las fórmulas clásicas (que asumen $h$ constante) introduciría un error sistemático, sobre todo en las zonas donde el paso es muy distinto del promedio. Por eso se usan las versiones adaptadas a paso variable.

2. **Coincidencia visual de los tres esquemas.** Los tres esquemas reproducen la misma tendencia global de la derivada. Para puntos del interior alejados de los extremos, sus curvas son prácticamente indistinguibles a escala global.

3. **Sesgo de progresiva y regresiva.** Como en el caso uniforme, la diferencia progresiva tiene un sesgo positivo (mira hacia adelante) y la regresiva un sesgo negativo (mira hacia atrás). Estos sesgos se hacen más evidentes en los tramos donde el paso $\Delta x$ es grande, ya que el error de truncamiento crece linealmente con el paso.

4. **Ventaja de la diferencia central no uniforme.** Al usar información de ambos lados ponderada por los pasos $h_1$ y $h_2$, esta fórmula recupera un error de orden $\mathcal{O}(h_1 h_2)$ (segundo orden combinado). Su error promedio frente a progresiva/regresiva es notablemente menor.

5. **Sensibilidad al paso local.** En las regiones donde $\Delta x$ es **muy pequeño** (puntos muy juntos), el error absoluto de cualquier esquema disminuye, pero se vuelve potencialmente más sensible al ruido numérico de los datos $y_i$. En cambio, donde $\Delta x$ es **grande**, el error de truncamiento domina, sobre todo para las diferencias de primer orden.

6. **Caso límite.** Si se forzara $h_1 = h_2 = h$ en la fórmula centrada no uniforme, se recupera exactamente la diferencia centrada clásica del inciso b). Esto confirma que la fórmula utilizada es una **generalización consistente** del esquema central visto en la conferencia, manteniendo el mismo orden de exactitud.

7. **Conclusión general.** Para nubes de puntos con espaciado no homogéneo es indispensable adaptar las fórmulas de diferencias finitas usando los pasos locales $h_1, h_2$. La diferencia central de tres puntos no uniformes ofrece el mejor compromiso entre simplicidad e exactitud entre los esquemas vistos en la conferencia.
