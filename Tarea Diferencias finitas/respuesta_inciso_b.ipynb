##### Esquemas de diferencias finitas aplicados

Dado que los puntos están **homogéneamente espaciados** con paso $h = x_{i+1} - x_i$ constante, podemos aplicar directamente los esquemas estándar vistos en la conferencia:

- **Diferencia progresiva** ($\mathcal{O}(h)$):
$$f'_n \approx \frac{f_{n+1} - f_n}{h}$$

- **Diferencia regresiva** ($\mathcal{O}(h)$):
$$f'_n \approx \frac{f_n - f_{n-1}}{h}$$

- **Diferencia central** ($\mathcal{O}(h^2)$):
$$f'_n \approx \frac{f_{n+1} - f_{n-1}}{2h}$$

- **Esquema de 4to orden** ($\mathcal{O}(h^4)$):
$$f'_n \approx \frac{f_{n-2} - 8\,f_{n-1} + 8\,f_{n+1} - f_{n+2}}{12\,h}$$

En cada esquema, los puntos extremos (donde no se dispone de vecinos suficientes) se rellenan con la diferencia progresiva o regresiva equivalente.


```python
def dif_progresiva(x, y):
    """Calcula derivada progresiva (forward difference) sobre puntos (x, y) con h constante"""
    h  = x[1] - x[0]
    df = np.zeros_like(y)
    # Diferencia progresiva en todos excepto el último punto
    df[:-1] = (y[1:] - y[:-1]) / h
    # En el último punto usamos diferencia regresiva
    df[-1]  = (y[-1] - y[-2]) / h
    return df, h


def dif_regresiva(x, y):
    """Calcula derivada regresiva (backward difference) sobre puntos (x, y) con h constante"""
    h  = x[1] - x[0]
    df = np.zeros_like(y)
    # Diferencia regresiva en todos excepto el primer punto
    df[1:] = (y[1:] - y[:-1]) / h
    # En el primer punto usamos diferencia progresiva
    df[0]  = (y[1]  - y[0])  / h
    return df, h


def dif_centrada(x, y):
    """Calcula derivada centrada de O(h^2) sobre puntos (x, y) con h constante"""
    h  = x[1] - x[0]
    df = np.zeros_like(y)
    # Diferencia central en los puntos interiores
    df[1:-1] = (y[2:] - y[:-2]) / (2*h)
    # Esquemas progresivo/regresivo en los extremos
    df[0]  = (y[1]  - y[0])  / h
    df[-1] = (y[-1] - y[-2]) / h
    return df, h


def dif_orden_4(x, y):
    """Calcula derivada centrada de O(h^4) sobre puntos (x, y) con h constante"""
    h  = x[1] - x[0]
    df = np.zeros_like(y)
    # Esquema centrado de cuarto orden (5 puntos)
    df[2:-2] = (y[:-4] - 8*y[1:-3] + 8*y[3:-1] - y[4:]) / (12*h)
    # En los dos primeros nodos: progresiva (O(h))
    df[0] = (y[1] - y[0]) / h
    df[1] = (y[2] - y[1]) / h
    # En los dos últimos nodos: regresiva (O(h))
    df[-2] = (y[-2] - y[-3]) / h
    df[-1] = (y[-1] - y[-2]) / h
    return df, h
```

##### Evaluación de la primera derivada

Aplicamos los cuatro esquemas a la nube de puntos $(x_i, y_i)$ y graficamos los resultados en un único panel para comparar visualmente.


```python
# Cálculo de la primera derivada con los cuatro esquemas
df_prog, h = dif_progresiva(x, y)
df_regr, _ = dif_regresiva(x, y)
df_cent, _ = dif_centrada(x, y)
df_ord4, _ = dif_orden_4(x, y)

p_print(fr'h = {h:.6f}')

# Graficación comparativa
fig, ax = plt.subplots(figsize=(7, 4), dpi=150)

ax.plot(x, df_prog, 'C1-',  linewidth=0.7, label='Progresiva  $\\mathcal{O}(h)$')
ax.plot(x, df_prog, 'C1.', markersize=3)

ax.plot(x, df_regr, 'C2-',  linewidth=0.7, label='Regresiva  $\\mathcal{O}(h)$')
ax.plot(x, df_regr, 'C2.', markersize=3)

ax.plot(x, df_cent, 'C3-',  linewidth=0.9, label='Centrada  $\\mathcal{O}(h^2)$')
ax.plot(x, df_cent, 'C3.', markersize=3)

ax.plot(x, df_ord4, 'C0-',  linewidth=0.9, label='4to orden  $\\mathcal{O}(h^4)$')
ax.plot(x, df_ord4, 'C0.', markersize=3)

ax.set_xlabel('x', fontsize=11)
ax.set_ylabel("$dy/dx$", fontsize=11)
ax.set_title("Primera derivada por diferencias finitas (espaciado homogéneo)", fontsize=11)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8, loc='best')
plt.tight_layout()
plt.show()
```

##### Comparación por separado

```python
# Comparación de cada esquema contra la derivada central (referencia)
fig, axes = plt.subplots(2, 2, figsize=(9, 6), dpi=150, sharex=True, sharey=True)

esquemas = [
    ('Progresiva  $\\mathcal{O}(h)$',  df_prog, 'C1'),
    ('Regresiva  $\\mathcal{O}(h)$',   df_regr, 'C2'),
    ('Centrada  $\\mathcal{O}(h^2)$',  df_cent, 'C3'),
    ('4to orden  $\\mathcal{O}(h^4)$', df_ord4, 'C0'),
]

for ax, (nombre, df, color) in zip(axes.flat, esquemas):
    ax.plot(x, df_ord4, 'k-', linewidth=0.8, alpha=0.5, label='Ref. 4to orden')
    ax.plot(x, df, f'{color}-', linewidth=0.7, label=nombre)
    ax.plot(x, df, f'{color}.', markersize=3)
    ax.set_title(nombre, fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=7, loc='best')

for ax in axes[-1, :]:
    ax.set_xlabel('x', fontsize=10)
for ax in axes[:, 0]:
    ax.set_ylabel('$dy/dx$', fontsize=10)

plt.tight_layout()
plt.show()
```

##### Comparación de errores entre esquemas

Como no se dispone de la función analítica $f(x)$, se utiliza el esquema de **cuarto orden** como referencia (su error de truncamiento es $\mathcal{O}(h^4)$, mucho menor que el de los demás esquemas). Se calcula entonces el error absoluto:

$$E_i = \left| f'_{\text{esquema}}(x_i) - f'_{\text{ord. 4}}(x_i) \right|$$

y se reporta el **error máximo** y el **error promedio** sobre los puntos interiores (descartando los extremos, donde los esquemas usan fórmulas distintas).


```python
# Tomamos como referencia el esquema de 4to orden (sobre los puntos interiores)
mask = np.zeros_like(x, dtype=bool)
mask[2:-2] = True  # puntos donde TODOS los esquemas usan su fórmula nominal

ref = df_ord4

errores = {
    'Progresiva': df_prog,
    'Regresiva':  df_regr,
    'Centrada':   df_cent,
}

# Encabezado
print(f"{'Esquema':<15}{'Error máx.':>16}{'Error prom.':>16}")
print('-' * 47)
for nombre, df in errores.items():
    err = np.abs(df[mask] - ref[mask])
    print(f"{nombre:<15}{err.max():>16.5e}{err.mean():>16.5e}")
```

##### Análisis de los resultados

A partir de los gráficos y de la tabla de errores se observan los siguientes comportamientos:

1. **Coincidencia visual de los esquemas en el interior.** Para puntos suficientemente alejados de los extremos, los cuatro esquemas producen curvas prácticamente superpuestas. Esto es esperable: con $h \approx 0.0202$ todos los esquemas convergen al valor real de la derivada, aunque con diferentes velocidades.

2. **Diferencias entre progresiva y regresiva.** Como ambos esquemas son de orden $\mathcal{O}(h)$, sus errores son del mismo orden, pero presentan un **sesgo** opuesto: la diferencia progresiva tiende a "adelantar" la derivada, mientras que la regresiva la "atrasa". Esto se traduce en pequeños desplazamientos de signo opuesto respecto a la derivada centrada.

3. **Ventaja de la diferencia central.** Al combinar información de ambos lados del punto, la diferencia centrada tiene error $\mathcal{O}(h^2)$, mejorando el orden en una unidad. Su error promedio frente al esquema de 4to orden es notablemente menor que el de las diferencias progresivas/regresivas.

4. **Esquema de cuarto orden como mejor estimador.** El esquema $\mathcal{O}(h^4)$ utiliza cinco puntos y, salvo en los nodos extremos (donde colapsa a un esquema de menor orden), produce la mejor estimación disponible de la derivada. Por esa razón se usó como referencia para cuantificar los errores de los demás esquemas.

5. **Comportamiento en los extremos.** Cerca de los bordes ($i = 0,\, 1,\, N-2,\, N-1$) el esquema de 4to orden cae a fórmulas progresivas/regresivas de menor orden, por lo que su precisión se degrada localmente. Esto es coherente con lo discutido para las derivadas de orden superior en la tarea previa.

6. **Conclusión general.** Para una nube de puntos homogéneamente espaciada, la diferencia centrada ofrece un balance excelente entre simplicidad y precisión, y el esquema de cuarto orden es la mejor opción cuando se requiere alta exactitud y se dispone de al menos dos vecinos a cada lado.
